```python
from typing import List
from collections import Counter
import heapq
from timeit import default_timer as timer
```

```python
def solution(arr, t):        
    # two-pointers technique
    l, r = 0, 1
    s = arr[l]
    while r<len(arr):
        if s==t:
            return [l, r-1]
        elif s<t:
            # if l==r, s==0, then s<t because t>0
            # this means that l>r never happens
            s+=arr[r]
            r+=1
        else: # s>t
            s-=arr[l]
            l+=1
    
    # now r == len(arr)
    while l<r:
        if s==t:
            return [l, r-1]
        elif s>t:
            s-=arr[l]
            l+=1
        else:
            return [-1, -1]
        
    return [-1, -1]
```

```python
for title, data, n, output in test_data:
    sol = solution(data, n)
    
    print(title)
    print(output)
    if sol==output:
        print(sol)
    else:
        print(f"{sol} <----")
    print()
```

```python
start = timer()

i = 0
while i<100000:
    for _, data, n, output in test_data:
        sol = solution(data, n)
    i+=1
    
end = timer()

print(f"# Elapsed time: {i}:{end-start}")
```
